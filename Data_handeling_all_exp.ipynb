```python
# Experiment: 1(c)
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

date_range = pd.date_range(start='2023-01-01', end='2025-12-31',freq="ME")
values = np.random.rand(len(date_range))*100
time_series_data = pd.DataFrame({'Data':date_range,'value':values})
plt.figure(figsize=(10,6))
plt.plot(time_series_data['Data'],time_series_data['value'],linestyle = '-')
plt.xlabel('Date',fontsize=12)
plt.ylabel('Value',fontsize=12)
plt.title('Time Series Data')
plt.grid(visible= True,linestyle ='--',color='gray')
plt.tight_layout()
plt.show()
```

```python
# Experiment 1(e)
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)
increasing = x
decreasing = 11 - x
fluctuating = np.array([2, 5, 3, 8, 6, 9, 4, 7, 3, 10])

plt.figure(figsize=(10, 6))
plt.plot(x, increasing, label='increasing trend', color='green', marker='o')
plt.plot(x, fluctuating, label='fluctuating trend', color='blue', marker='o')

plt.title("line plot with different type of changes", fontsize=16)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.grid(visible=True, linestyle='--', color='gray')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Experiment 2(c):
import matplotlib.pyplot as plt

categories = ['red', 'blue', 'green', 'yellow', 'purple']
frequencies = [15, 20, 10, 18, 12]

plt.figure(figsize=(8, 5))
colors = categories
bars = plt.bar(categories, frequencies, color=colors)

plt.bar_label(bars, labels=categories, label_type='edge', color='black')
plt.title('Frequence bar plt', fontsize=14)
plt.xlabel('colors', fontsize=12)
plt.ylabel('Frequencies', fontsize=12)
plt.show()
```

```python
# Experiment 3(a) :
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'months' : ['jan','feb','mar','apr','may'],
    'sales-A' : [200,250,300,280,350],
    'sales-B' : [180,220,260,240,310]
}
df = pd.DataFrame(data)
x = df.set_index('months')

x.plot(kind='line',figsize=(8,5),style=['--','-.'],color=['orange','blue'],title='monthly sales comparison',markerfacecolor='black',linewidth=2)
plt.xlabel('months')
plt.ylabel('sales')
plt.grid(True,linestyle = '-',linewidth=2)
plt.show()
```

```python
#Experiment 3(b):
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'months': ['jan', 'feb', 'mar', 'apr', 'may'],
    'sales-A': [200, 230, 300, 280, 350],
    'sales-B': [40, 60, 70, 80, 50]
}
df = pd.DataFrame(data)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar(df['months'], df['sales-A'], label='Sales-A')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Profit')
axes[0].grid(True, linestyle='--', alpha=0.6, axis='y')

axes[1].plot(df['months'], df['sales-B'], marker='o', label='Sales-B', color='orange')
axes[1].set_title('profit-B')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Profit')
axes[1].grid(True, linestyle=' ', alpha=0.06, linewidth=2, color='darkblue')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


penguins = sns.load_dataset("penguins")
fmri = sns.load_dataset("fmri")


sns.relplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    kind="scatter"
)
plt.title("Penguins: Flipper Length vs Body Mass")
plt.show()

sns.relplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    size="bill_length_mm",
    style="sex",
    kind="scatter"
)
plt.title("Penguins: Flipper Length vs Body Mass (Categorized)")
plt.show()


sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    style="region",
    kind="line"
)
plt.title("fMRI: Signal Over Time")
plt.show()


sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    col="region",
    kind="line"
)
plt.tight_layout
plt.show()
```
